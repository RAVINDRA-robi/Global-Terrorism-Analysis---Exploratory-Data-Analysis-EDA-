MAIN **PART**

1.Data loading

````
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
````

````
df = pd.read_csv("/content/Global Terrorism Data.csv" , encoding='latin1')
df.head()
````

2.Data Understanding

````
df.shape
df.columns
df.info()
df.describe()
````

3.data cleaning

````
df.fillna(0, inplace=True)
````

**HYPOTHESIS: **

```
Terrorist attacks increased over time
Certain countries are highly affected
Bombing is the most common attack type
Casualties increased in recent years
```



4.Visualization(Attacks per year)

````
df['iyear'].value_counts().sort_index().plot(kind='line')
plt.title("Attacks per Year")
plt.xlabel("Year")
plt.ylabel("Number of Attacks")
plt.show()
````

5.Visualization(Top countries)

````
df['country_txt'].value_counts().head(10).plot(kind='bar')
plt.title("top 15 countries with most attacks")
plt.show()
````

6.Visualization(Weapon Types)

````
df['weaptype1_txt'].value_counts().head(15).plot(kind='bar')
plt.title("Weapon Types")
plt.show()
````

7.Visualization(Casualties)

````
df.groupby('iyear')['nkill'].sum().plot(kind='line')
plt.title("Total Casualities per Year")
plt.show()
````

# **CONCLUSION**

```
The analysis shows that terrorism has increased significantly over time, especially after 2000.
Certain countries like Iraq and Afghanistan are highly affected.
Bombing and explosive attacks are the most common types of attacks.

These insights can help governments and organizations to focus on high-risk regions and improve security measures.
